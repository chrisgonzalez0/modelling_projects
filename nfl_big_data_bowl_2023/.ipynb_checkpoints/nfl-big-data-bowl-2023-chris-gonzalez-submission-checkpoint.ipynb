# Using CNNs and Integrated Gradients to Evaluate Linemen (Offense and Defense) Contribution to Expected Yards

# Introduction

The game of football (as we all know and love) might be the most complex system of interactions and dependencies between individual players compared to all other Major Sports. Sports like basketball, soccer, baseball, and hockey, while team based, are also typically more fluid in that offensive and defensive players have more in common similar technical skills with each other, more similar physical attributes, and have to both play offense and defense. Individual players in other team sports are more 'similar to' one another and thus more easily comparable to each other given statistics collected about them.

One way to think about this is to use an example such as the NBA. There are a finite set of skills that players have (to varying degrees) and physical attributes (to varying degrees). These skills/attributes include dribbling, shooting, passing, rebounding, speed, and players range mostly in height physically. While some players might be a bit more specialized, there is a lot of cross over between positions and these attributes.

When we look at football, however, the varying degrees of position groups and how they co-depend on one another varies greatly. In football, an Offensive linemen needs not have any of the skillsets assigned to that of a quarterback or skill position groups (WR/HB). An offensive linemen simply needs to generate enough Force and fill the gaps to keep potential pass rushers at bay and buy the quarterback time to throw the ball comfortably. They also have less control over other parts of the game given this specialization and number of players on the field. In a similar vein, other positional groups can be described as such. 

What makes football unique (and also the most complex) is that it faces the same question as the classic 'chicken or the egg' paradox. In football, we can observe the success of a play if we measure something like 'yards gained (or lost),' but like the chicken-egg paradox we can't really unscramble what caused what. Was it the the offensive line protecting the quarterback that resulted in the outcome? Did the quarterback get the ball out quickly or was he savvy enough to find the receiver? Did the receivers very quickly get seperation from the corners? Did the corner not do a good enough job keeping with the receiver? And on it can go as we attempt to find reasons for a particular play's success. Even if we generalize the potential reasons for a play's success, how can we assign a value to each one? Positional groups in football depend on each other much like the Chicken and the Egg question poses.

In truth (and like most analysis), the most likely reason is a combination of all of these factors. If a play gains 20 yards, we should most likely attribute some percentage of that 20 yds to the performance of the offensive line, some percent to the quarterback, some to the skill positions, some to the defense, etc. This analysis using Convolutional Neural Networks and using X-Y coordinates, the weight of individual players, and their acceleration at each frame of a particular play will attempt to do so in the following Notebook.

# The Data and Data Clean-Up

This analysis uses the NFL Next Gen Stats data provided by the competition and makes the following transformations to it given my own local computing resources and access to RAM on my 2015 MacBook Air (a rather old laptop). A future iteration on this analysis might better utilize more advanced technical resources and improve upon this work.

1. Because I am choosing to build a convolutional neural network, I decided to rotate all x, y coordinates where the playDirection was marked as 'left' to move 'right.' If the play is moving to your left, just imagine you are sitting on the opposite side of the stadium and now it moves right, this is the same idea. This has no effect on the actual play or data but this makes my convolutional neural network learn 'more quickly' given this standardization.

2. The resulting x, y coordinates after the rotation described in 1) were then rounded to the nearest integer and I removed any x,y coordinates from the player tracking data that fell outside of the range between (0-120, 0-53). This standardization meant that I could now build image tensors with a pixel size of 54 by 121. I could have attempted to use the non-rounded X-Y coordinates but that would have resulted in images of size 5400 by 12100 which is much to large for resources at my disposal.

3. I removed any plays that had a penalty occur from the analysis given that these types of plays (like holding) are more nuanced and not capturable in detail by simple x, y coordinates.

4. Finally, in an attempt to help the Convolutional Neural Network distinguish between the offense, defense, and position of the football, I created a tensor (or collection of datasets) with the shape [3,54,121] corresponding to [ (offense, defense, or football), (Y length of the football field), (X length of the football field) ] for a single frame of any given play in a game provided in the data. Also, rather than use a simple binary value for whether the player was at position x,y, I used their normalized (between 0.1 and 1) Weight times Acceleration at that given moment as a proxy for Force Generated by the player.See the Appendix (under get_tensors) for how this function creates these tensors and the image below for a visual of how the tensor looks like. Notice the 3 seperate datasets that correspond to offense, defense, and football, and their normalized (0.1 to 1 values) along where they are on the field on the x,y axis.

# What is a Convolutional Neural Network and How are we using it here? 

Convolutional Neural Neural Networks (or CNN's for short) are a type of Data Science Model typically used for Image Classification. A CNN with enough data to learn from can take a picture of a dog and correctly say that it is a picture of a dog, or it can take a single image with multiple objects and label all of them. All of this depends on the architecture of the network and the data that it has to learn from. They utilize functions that are called Convolutions which take groups of pixels and applies weights to them to generalize an image. 

Naturally and based on what we've described so far, we can see how we might be able to leverage a CNN for our analysis. If you recall from the section 'The Data and Data Clean-Up,' we've constructed a data structure that looks a lot like an image. For the purposes of this analysis, however, instead of having images that either have values of Red, Green, or Blue overlayed on top of each other to produce an image, our images are offense, football, and defense positions with the values of the x,y coordinates corresponding to a normalized value of the player's weight*acceleration (or Force as I call it) if the player is at that exact location.  

Additionally, instead of trying to predict whether our set of images corresponding to the play is an object, I instead predict how many yards we should expect given the x,y coordinates of the players, ball, and their normalized weights/accelerations at the given frame. Thus, the input is the set of offense, football, and defense images and the prediction is how many yards the CNN predicts will be the result of the play using only that single frame.

# Why are we using a Convolutional Neural Network?

Because (as described in the intro of this Notebook) the game of football is an extremely complex game that requires different positional groups to support one another in different parts of the field, we need a model that is best able to recognize the nuances described previously where one positional group affects the other and controls for these effects. Using pre-built Python packages that help to 'Explain' Neural Network models, we can also 'highlight' (using a concept known as Integrated Gradients) which parts of the image are the 'reasons' as to why we are making a particular prediction. Consider the following image and high level explanation taken from this article: https://distill.pub/2020/attribution-baselines/



Integrated Gradients at a high-level takes a 'baseline' image (in our case one that looks like the image) and gradually changes that baseline image to look closer to the original image (using an arbitrarily chosen step size) and computes how much each pixel 'contributes' towards our prediction. So for a step size of 20, it would generate 20 images such that each iteration look more like the orignal image and extrapolates how much each pixel contributed to the prediction. In Data Science, it is similar to doing Partial Effects Analysis on the Inputs of a Regression model but instead for a Neural Network. 

In essence, what I am seeking to accomplish is similar to the image above. I want to first build a CNN that can predict expected yards accurately given a single frame of a play, then highlight the parts of the field such that they are contributing to that final outcome (like the bird image above). The Appendix has code that 1) Describes the CNN architecture chosen, 2) the training of the CNN, and 3) subsequent analysis of individual players, most notable offensive and defensive linemen given the Kaggle Competition.

# Analysis of Offensive and Defensive Linemen

Now that we've 

1) Built a CNN to modestly train the model ( Training MSE Loss: 2, Testing MSE Loss: 10 )

2) Used Integrated Gradients to give 'Attributions' to each part of the field (every pixel)

3) And visualized our pixel attributions of our model overlayed on actual plays

We can now do a high level analysis to summarize each player who had a PFF Role of 'Pass Rush' or 'Pass Block' and their average contribution to our Predicted Yards determined by the CNN. The subsequent tables show a potentially new and novel way to assign value to Pass Blockers and Pass Rushers given everything happening on the field of play at any given frame. 

This high level analysis will define 2 different metrics to analyze our Pass Rushers and Blockers: 

1) Player Direct Contribution: We will take the Pixel Attribution that the player is currently on. For example, if a player is located at X coordinate=33 and y coordinate=45, we will take the Pixel Attribution at that exact location in the frame and assign it to the player as his direct contribution to the outcome of the play. 

2) Player Surrounding Contribution: We will take the Pixel Attribution that the player is currently on AND the surrounding 1 pixels. For example, if a player is on x coordinate=33 and y coordinate=45, we will sum the Pixel Attributions in the square defined by x in [32 to 34] and y in [44 to 46].

We will then aggregate these values across every play and frame and present the results below as measured by 'Best Individual' (Direct Contribution) and 'Best Gravitational' (Surrounding Contribution) Pass Blockers and Pass Rushers.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
pd.options.mode.chained_assignment = None
import os
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from textwrap import wrap
from IPython.display import HTML

dir=''

games=pd.read_csv('/kaggle/input/nfl-big-data-bowl-2023/games.csv')
scouting=pd.read_csv('/kaggle/input/nfl-big-data-bowl-2023/pffScoutingData.csv')
players=pd.read_csv('/kaggle/input/nfl-big-data-bowl-2023/players.csv')
plays=pd.read_csv('/kaggle/input/nfl-big-data-bowl-2023/plays.csv')

plays=plays.loc[plays.foulName1.isna(),:]
plays.loc[:,'normed_playResult']=plays.loc[:,'playResult']
y_means=plays.normed_playResult.mean()
y_std=plays.normed_playResult.std()
plays.loc[:,'normed_playResult']=(plays.loc[:,'normed_playResult']-y_means)/y_std


week=pd.read_csv('/kaggle/input/nfl-big-data-bowl-2023/week1.csv')
for i in range(2,9):
    week=pd.concat([week,
                    pd.read_csv('/kaggle/input/nfl-big-data-bowl-2023/week'+str(i)+'.csv')],
                   axis=0)
week=week.loc[(week.x >= 0) & (week.x <= 120) & (week.y >= 0) & (week.y <= 53.3),:]

colors=pd.read_csv('/kaggle/input/nflbigdatabowl/team_colors.csv')

week=pd.merge(week,colors,'left',left_on='team',right_on='team')
week=pd.merge(week,players.loc[:,['nflId','weight','officialPosition']],'left',left_on='nflId', right_on='nflId')
week.loc[week.officialPosition.isna(),'officialPosition']=''
week=pd.merge(week,scouting.loc[:,['nflId','gameId','playId','pff_positionLinedUp','pff_role']],'left',
              left_on=['nflId','gameId','playId'], 
              right_on=['nflId','gameId','playId'])
week.loc[week.pff_positionLinedUp.isna(),'pff_positionLinedUp']=''
week.loc[week.pff_role.isna(),'pff_role']=''

## make all plays go the same direction
week.x.loc[week.playDirection=='left']=120-week.x.loc[week.playDirection=='left']
week.y.loc[week.playDirection=='left']=53.3-week.y.loc[week.playDirection=='left']
week.dir.loc[week.playDirection=='left']=((week.dir.loc[week.playDirection=='left']-360)-180)%360

week.x=round(week.x,0)
week.y=round(week.y,0)


print('Weekly Frame by Frame Data is this size in [rows, columns]: '+ str(week.shape))
week.head()
```

Section 2: Defining a Function to Animate a Play

```python

def animate(gameid,playid):
    fig, ax = plt.subplots(dpi=2500/16)
    im = plt.imread('/kaggle/input/nflbigdatabowl/football_field.jpg')
    _ = plt.imshow(im,cmap='gray',zorder=1,origin='lower',extent=[0,120,0,54])
    
    s=ax.scatter([],[])
    
    ax.set_title("\n".join(wrap(
        plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'playDescription'].item()
                                ,50)),fontsize=6)
    ax.set_xlabel("\n".join(wrap(        
        str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'down'].item())+' and '
        +str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'yardsToGo'].item())+' at '
        +str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'yardlineSide'].item())+' '
        +str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'yardlineNumber'].item())        
                                ,50)),fontsize=6
                 )

    ax.set_ylabel("\n".join(wrap(        
        'Offense: '+str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'possessionTeam'].item())+'\n'
        +'Defense: '+str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'defensiveTeam'].item())
                                ,50)),fontsize=6
                 )
    
    dataset=week.loc[(week.playId==playid) & (week.gameId==gameid),:]
    #dataset.loc[:,'x']=dataset.loc[:,'x']*883/120
    #dataset.loc[:,'y']=dataset.loc[:,'y']*359/53.3
    dataset.loc[:,'sizes']=90
    dataset.loc[dataset.team=='football','sizes']=7
    
    dataset['dir1']=np.sin(dataset.dir*np.pi/180)
    dataset['dir2']=np.cos(dataset.dir*np.pi/180)
    qr = ax.quiver(dataset.loc[dataset.frameId==(1),'x'], 
                   dataset.loc[dataset.frameId==(1),'y'], 
                   dataset.loc[dataset.frameId==(1),'dir1'], 
                   dataset.loc[dataset.frameId==(1),'dir2'], 
                   color='black')
        
    frames=dataset.frameId.unique()
    fe=dataset.loc[:,['frameId','event']].drop_duplicates()  ## frame by event that occurred
    fe=fe.reset_index(drop=True)
    
    title = ax.text(0.5,0.85, "", bbox={'facecolor':'w', 'alpha':0.5, 'pad':5},
                transform=ax.transAxes, ha="center")
    
    init_df=dataset.loc[dataset.frameId==1,:].reset_index(drop=True)
    
    annotation = [ax.annotate(
        np.array(init_df.loc[j,'officialPosition']).item(), 
        xy=(np.array(init_df.loc[j,['x','y'] ])), va='center', ha='center',fontsize=4,color='white') for j in range(init_df.shape[0])]

    def init():
        #plt.xlim(0, 120)
        #plt.ylim(0, 53.3) 
        ax.axis([min(dataset.x), max(dataset.x), 0 , 54])        
        s=ax.scatter([],[])        
        return s,

    def plot_play(i):
        frame_df=np.array(dataset.loc[dataset.frameId==(i),['x','y']])
        s.set_offsets(dataset.loc[dataset.frameId==(i),['x','y']])
        s.set_color(dataset.loc[dataset.frameId==(i), 'color_1' ])
        s.set_edgecolors(dataset.loc[dataset.frameId==(i), 'color_2' ])
        s.set_sizes(dataset.loc[dataset.frameId==(i), 'sizes' ])
        
        [annotation[j].set_position(tuple(frame_df[j]) ) for j in range(len(frame_df))]
        event=fe.loc[fe.frameId==(i),'event'].item()        
        if event!='None':
            title.set_text(fe.loc[fe.frameId==(i),'event'].item() )
            
        qr.set_offsets(dataset.loc[dataset.frameId==(i),['x','y']])
        qr.set_UVC( dataset.loc[dataset.frameId==(i),'dir1'] ,
                    dataset.loc[dataset.frameId==(i),'dir2'] )
        qr.set_color(dataset.loc[dataset.frameId==(i),'color_1'])
        return s,title,qr,

    plt.close('all')
    return animation.FuncAnimation(fig,
                                  plot_play, 
                                  frames=frames,
                                  blit=True,
                                  init_func=init,
                                  interval=350
                                  )
    
```

Section 3: Defining the CNN Class and Loading the Trained Convolutional Neural Network

(Note: Training was done using my own laptop locally, See Section 7 of the Appendix)

```python
## Import Integrated Gradients Package
from captum.attr import IntegratedGradients

## Helper functions for Convolutional NN
import torch
from torch.nn import Module
from torch.nn import Conv2d
from torch.nn import Linear
from torch.nn import MaxPool2d
from torch.nn import ReLU
from torch.nn import LogSoftmax
from torch import flatten
from torch import squeeze, unsqueeze


## We use the Architecture of one of the most standard convoluational neural networks but in it's untrained form. 
## Code in the appendix and in the input file nfl_functions.py demonstrates how gradient descent was performed 
## to train the model
class LeNet(Module):
	def __init__(self, numChannels, classes):
		# call the parent constructor
		super(LeNet, self).__init__()
		# initialize first set of CONV => RELU => POOL layers
		self.conv1 = Conv2d(in_channels=numChannels, out_channels=20,
			kernel_size=(2, 2))
		self.relu1 = ReLU()
		self.maxpool1 = MaxPool2d(kernel_size=(2, 2), stride=(2, 2))
		# initialize second set of CONV => RELU => POOL layers
		self.conv2 = Conv2d(in_channels=20, out_channels=40,
			kernel_size=(2, 2))
		self.relu2 = ReLU()
		self.maxpool2 = MaxPool2d(kernel_size=(2, 2), stride=(3, 3))
		# initialize first (and only) set of FC => RELU layers
		self.fc1 = Linear(in_features=6400, out_features=500)
		self.relu3 = ReLU()
		# initialize our softmax classifier
		self.fc2 = Linear(in_features=500, out_features=classes)
		self.logSoftmax = LogSoftmax(dim=1)
        
	def forward(self, x):
		# pass the input through our first set of CONV => RELU =>
		# POOL layers
		x = self.conv1(x)
		x = self.relu1(x)
		x = self.maxpool1(x)
		# pass the output from the previous layer through the second
		# set of CONV => RELU => POOL layers
		x = self.conv2(x)
		x = self.relu2(x)
		x = self.maxpool2(x)
		# flatten the output from the previous layer and pass it
		# through our only set of FC => RELU layers
		x = flatten(x)
		x = self.fc1(x)
		x = self.relu3(x)
		# predictions
		x = self.fc2(x)
		output = x
		# return the output predictions
		return output        

## This is the final trained model saved from nfl_functions.py
## and trained locally on my Macbook CPU
model=torch.load('/kaggle/input/nflbigdatabowlfiles2/lenet_v2.pt')
model.eval()
```

Section 4: Define a Function Called get_tensors that restructures frames into tensors of size [3,54,121] corresponding to [ (offense,football,defense) , (Y Coordinates of Field) , (X Coordinates of Field) ]

```python
## We define a function called get_tensors 
def get_tensors(gameId,playId,frameId):
    
    #gameId=2021101703
    #playId=1779
    #frameId=53
    
    temp=week.loc[(week.gameId==gameId) & (week.playId==playId) & (week.frameId==frameId),:]
    temp.loc[temp.team=='football','weight']=1
    temp.loc[:,'force']=temp.weight*temp.a
    temp.loc[temp.team=='football','force']=1
    #temp.loc[temp.team!='football','force']=(temp.loc[temp.team!='football','force']-temp.force[temp.team!='football'].mean())/temp.force[temp.team!='football'].std()
    temp.loc[temp.team!='football','force']=0.9*(temp.loc[temp.team!='football','force']-(temp.force[temp.team!='football'].min()+0.1) )/( temp.force[temp.team!='football'].max() - (temp.force[temp.team!='football'].min()+0.1) ) + 0.1    
    
    layers=['offense','football','defense']
    x_data=torch.zeros([3,54,121])
    
    for j in range(len(layers)):
        tens_temp=x_data[j]
        
        if layers[j]=='offense':
            temp_j=temp.loc[temp.team==plays.possessionTeam[(plays.gameId==gameId) & (plays.playId==playId)].item(),:]
            temp_j=temp_j.reset_index(drop=True)
        if layers[j]=='defense':
            temp_j=temp.loc[temp.team==plays.defensiveTeam[(plays.gameId==gameId) & (plays.playId==playId)].item(),:]
            temp_j=temp_j.reset_index(drop=True)        
        if layers[j]=='football':
            temp_j=temp.loc[temp.team=='football',:]
            temp_j=temp_j.reset_index(drop=True)
        
        for i in range(temp_j.shape[0]):
            tens_temp[int(temp_j.y[i].item()),int(temp_j.x[i].item())]=temp_j.force[i]
            #print(tens_temp[int(temp_j.y[i].item()*10),int(temp_j.x[i].item()*10)])
        
        x_data[j]=tens_temp
    
    y_data=torch.tensor([plays.normed_playResult[(plays.gameId==gameId) & (plays.playId==playId)].item()])
    
    return x_data,y_data

##############################################
```

```python
from captum.attr import IntegratedGradients
ig = IntegratedGradients(model)

def animate_with_heat(gameid,playid):
    fig, ax = plt.subplots(dpi=2500/16)
    im = plt.imread('/kaggle/input/nflbigdatabowl/football_field.jpg')
    _ = plt.imshow(im,cmap='gray',zorder=1,origin='lower',extent=[0,120,0,54])

    heat=ax.scatter([],[],marker='s',c=[],cmap='RdYlGn',s=60)
    s=ax.scatter([],[])

    ax.set_title("\n".join(wrap(
        plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'playDescription'].item()
                                ,50)),fontsize=6)
    ax.set_xlabel("\n".join(wrap(        
        str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'down'].item())+' and '
        +str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'yardsToGo'].item())+' at '
        +str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'yardlineSide'].item())+' '
        +str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'yardlineNumber'].item())        
                                ,50)),fontsize=6
                 )

    ax.set_ylabel("\n".join(wrap(        
        'Offense: '+str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'possessionTeam'].item())+'\n'
        +'Defense: '+str(plays.loc[(plays.gameId==gameid) & (plays.playId==playid), 'defensiveTeam'].item())
                                ,50)),fontsize=6
                 )
    
    dataset=week.loc[(week.playId==playid) & (week.gameId==gameid),:]
    dataset.loc[:,'sizes']=60
    dataset.loc[dataset.team=='football','sizes']=7
    
    dataset['dir1']=np.sin(dataset.dir*np.pi/180)
    dataset['dir2']=np.cos(dataset.dir*np.pi/180)
    qr = ax.quiver(dataset.loc[dataset.frameId==(1),'x'], 
                   dataset.loc[dataset.frameId==(1),'y'], 
                   dataset.loc[dataset.frameId==(1),'dir1'], 
                   dataset.loc[dataset.frameId==(1),'dir2'], 
                   color='black')
        
    frames=dataset.frameId.unique()
    fe=dataset.loc[:,['frameId','event']].drop_duplicates()  ## frame by event that occurred
    fe=fe.reset_index(drop=True)
    
    title = ax.text(0.5,0.85, "", bbox={'facecolor':'w', 'alpha':0.5, 'pad':5},
                transform=ax.transAxes, ha="center")
    pred_label = ax.text(0.9,0.0, "", bbox={'facecolor':'w', 'alpha':0.5, 'pad':5},
                transform=ax.transAxes, ha="center",fontsize='x-small')

    init_df=dataset.loc[dataset.frameId==1,:].reset_index(drop=True)
    
    annotation = [ax.annotate(
        np.array(init_df.loc[j,'officialPosition']).item(), 
        xy=(np.array(init_df.loc[j,['x','y'] ])), va='center', ha='center',fontsize=4,color='white') for j in range(init_df.shape[0])]

    def init():
        ax.axis([min(dataset.x), max(dataset.x), 0 , 54])        
        s=ax.scatter([],[])     
        heat=ax.scatter([],[],marker='s',c=[],cmap='RdYlGn')
        return s,heat,

    def plot_play(i):

        x,y=get_tensors(dataset.gameId.unique().item(),dataset.playId.unique().item(),i)
        pred=model(x)
        pred_label.set_text('Predicted Yds: '+str(int(pred.item()*y_std+y_means)))
        
        baseimage=x+(torch.ones(3,54,121)*0.01)
        baseimage=baseimage.unsqueeze(0)
        imp_m = ig.attribute(inputs=x,
                                  baselines=baseimage,
                                  n_steps=50,internal_batch_size=3)

        imp_m=imp_m.squeeze(0)
        imp_m=torch.sum(imp_m,dim=0)        
        count=0
        temp=pd.DataFrame(columns=['y','x','c'])

        for k in range(imp_m.shape[0]):
            for l in range(int(min(dataset.x)), int(max(dataset.x))):
                temp.loc[count]=[int(k),int(l),imp_m[k,l].item()]
                count=count+1
                
        temp.loc[:,'c']=np.exp(temp.loc[:,'c'])/(1+np.exp(temp.loc[:,'c']))
        temp['sizes']=60
        
        heat.set_offsets(temp.loc[:,['x','y'] ] )
        heat.set_array( temp.loc[:,'c'] )
        heat.set_sizes( temp.loc[:,'sizes'] )

        frame_df=np.array(dataset.loc[dataset.frameId==(i),['x','y']])
        s.set_offsets(dataset.loc[dataset.frameId==(i),['x','y']])
        s.set_color(dataset.loc[dataset.frameId==(i), 'color_1' ])
        s.set_edgecolors(dataset.loc[dataset.frameId==(i), 'color_2' ])
        s.set_sizes(dataset.loc[dataset.frameId==(i), 'sizes' ])
        
        [annotation[j].set_position(tuple(frame_df[j]) ) for j in range(len(frame_df))]
        event=fe.loc[fe.frameId==(i),'event'].item()        
        if event!='None':
            title.set_text(fe.loc[fe.frameId==(i),'event'].item() )
            
        qr.set_offsets(dataset.loc[dataset.frameId==(i),['x','y']])
        qr.set_UVC( dataset.loc[dataset.frameId==(i),'dir1'] ,
                    dataset.loc[dataset.frameId==(i),'dir2'] )
        qr.set_color(dataset.loc[dataset.frameId==(i),'color_1'])
        
        return s,title,qr,heat,pred_label,
    plt.close('all')

    return animation.FuncAnimation(fig,
                                  plot_play, 
                                  frames=frames,
                                  blit=True,
                                  init_func=init,
                                  interval=350
                                  )
    
```

```python
all_plays=pd.concat(
    [pd.read_csv('/kaggle/input/nflbigdatabowl-files/test_frames.csv').reset_index(drop=True),
    pd.read_csv('/kaggle/input/nflbigdatabowl-files/train_frames.csv').reset_index(drop=True)],
    axis=0
)
all_plays=all_plays.reset_index(drop=True)
print(all_plays.shape)

week['predicted_yds']=0
week['player_direct_contribution']=0
week['player_surrounding_contribution']=0
week['sum_vals']=0

model.eval()
for i in range(all_plays.shape[0]):
    if (i+1)%1000==0:
        print(i)
        week.to_csv('all_week_analysis.csv',index=False)
        
    x,y=get_tensors(all_plays.gameId[i],all_plays.playId[i],all_plays.frameId[i])
    pred=model(x)
    
    baseimage=x+(torch.ones(3,54,121)*0.01)
    baseimage=baseimage.unsqueeze(0)    
    imp_m = ig.attribute(inputs=x,
                         baselines=baseimage,
                         n_steps=50,internal_batch_size=3)
    imp_m=imp_m.squeeze(0)
    imp_m=torch.sum(imp_m,dim=0)  
        
    week.loc[(week.gameId==all_plays.gameId[i]) & (week.playId==all_plays.playId[i]) & (week.frameId==all_plays.frameId[i]),'predicted_yds']=pred.item()*y_std+y_means
    week.loc[(week.gameId==all_plays.gameId[i]) & (week.playId==all_plays.playId[i]) & (week.frameId==all_plays.frameId[i]),'sum_vals']=abs(imp_m).sum().item()
    
    temp=week.loc[(week.gameId==all_plays.gameId[i]) & (week.playId==all_plays.playId[i]) & (week.frameId==all_plays.frameId[i]),['x','y','nflId']]
    temp=temp.loc[~(temp.nflId.isna()),:]
    temp=temp.reset_index(drop=True)
    for j in range(temp.shape[0]):
        direct_value=imp_m[int(temp.y[j]),int(temp.x[j])].item()
        surrounding_value=imp_m[int(temp.y[j]-1):int(temp.y[j]+2),int(temp.x[j]-1):int(temp.x[j]+2)].sum().item()
        
        week.loc[(week.gameId==all_plays.gameId[i]) & (week.playId==all_plays.playId[i]) & (week.frameId==all_plays.frameId[i]) 
                & (week.nflId==temp.nflId[j]), 'player_direct_contribution']=direct_value

        week.loc[(week.gameId==all_plays.gameId[i]) & (week.playId==all_plays.playId[i]) & (week.frameId==all_plays.frameId[i]) 
                & (week.nflId==temp.nflId[j]), 'player_surrounding_contribution']=surrounding_value
            
## this is the raw file with data for aggregate analysis
week.to_csv('all_week_analysis.csv',index=False)    


## Build Summary Data Set
final['direct_percent_attr']=0
final['surr_percent_attr']=0
final.direct_percent_attr=(final.player_direct_contribution/final.sum_vals)*final.predicted_yds
final.surr_percent_attr=(final.player_surrounding_contribution/final.sum_vals)*final.predicted_yds

final=pd.merge(final,scouting.loc[:,['nflId','gameId','playId','pff_role']],'left',
              left_on=['nflId','gameId','playId'], 
              right_on=['nflId','gameId','playId'])


means=final.groupby(['pff_role','nflId'],as_index=False)['direct_percent_attr','surr_percent_attr'].mean()
lengths=final.groupby(['pff_role','nflId'],as_index=False).size()
means=pd.merge(means,lengths,'left',left_on=['pff_role','nflId'],right_on=['pff_role','nflId']).reset_index(drop=True)

means=pd.merge(means,players.loc[:,['nflId','displayName']])

## Only look at Pass Blockers and Rushers and those with over 800 Frames in the data
means=means.loc[(means.pff_role=='Pass Block') | (means.pff_role=='Pass Rush'),:]
means=means.loc[means['size'] > 800,:]
```
